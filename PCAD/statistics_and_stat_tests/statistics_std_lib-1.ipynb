```python
from random import random
import statistics as stat

nums = []
more_nums = []

for i in range(10):
    nums.append(round(random(),2))

for i in range(10):
    more_nums.append(round(random(),2))
```

```python
nums
```

```python
more_nums
```

```python
stat.median_low(nums)
```

```python
stat.median_high(nums)
```

```python
stat.variance(nums)
```

```python
liner_corr = stat.correlation(nums, more_nums)
ranked_corr= stat.correlation(nums, more_nums, method='ranked')

print(liner_corr, ranked_corr)
```
