**NumPy**

NumPy is used for numerical computing — it provides powerful N-dimensional arrays and mathematical functions to work with large datasets efficiently.

- Fast mathematical operations on large data

- Replaces slow Python lists with efficient arrays

- Foundation for other libraries like Pandas, Scikit-learn, etc.

```python
#Create numpy array
import numpy as np
arr = np.array([10,20,30,40,50])
#arr = np.array([10,20,30,40,50,'abc'])
print(arr)
print(arr[0]) #follows indexing
print(type(arr))
print(arr.dtype)
```

```python
#arange(start,stop,step)
arr = np.arange(10,50)
print(arr)
```

```python
#Random array
arr = np.random.randint(10,100,5) #(start,stop,size)
print(arr)
```

```python
#2D array [rows,columns]
arr=np.array([[10,20,30,100],[40,50,60,200],[70,80,90,300]])
print(arr)
print(arr.ndim)
print(arr.shape)
print(arr[2][1])
```

```python
#Matrix Multiplication
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[10,11],[20,21],[30,31]])
print(arr1)  #2X3
print(arr2) #3X2
arr3=np.dot(arr1,arr2)
print(arr3)
```

```python
#Reshaping
arr=np.array([40,50,60,70,80,90])
#Change it to 2x3
arr_new=arr.reshape(3,2)
print(arr_new)
arr_new=arr.reshape(2,2) #Error:can not change the size
print(arr_new)
```

```python
print(arr_new) #3x2
print(np.transpose(arr_new)) #2x3
print(arr_new.T)
```

```python
#Boardcasting of Airthmetic Operations

lst=[10,20,30]
print(lst*5)

arr=np.array([10,20,30])
print(arr*5)
```

```python
#Stats Functions
sales=np.array([456,123,789,489,453])
print("Total Sales:",sales.sum())
print("Min Sales:",sales.min())
print("Max Sales:",sales.max())
print("Mean Sales:",sales.mean())
print("Median Sales:",np.median(sales))
print("Standard Deviation:",sales.std())
print("Variance:",sales.var())
```

**Pandas- Data Manupulation and analysis library**

Pandas is used for data manipulation and analysis.
It provides two main structures:

- Series (1D data)

- DataFrame (2D tabular data like Excel)

**Why use Pandas?**

- Easy to load, clean, and analyze data

- Handles CSV, Excel, SQL, JSON, etc.

- Integrates seamlessly with NumPy and Matplotlib

```python
#Import Data into Dataframe
import pandas as pd
df=pd.read_csv("/content/Social_Network_Ads.csv")
print(df)
```

```python
print(df.head(10))
```

```python
print(df.tail())
```

```python
print(df.info())
```

```python
#Stats Summery
print(df.describe())
print(df['Age'].describe())
```

Selections/Updates: specific rows,columns

```python
#Select specific columns
print(df[['Age','EstimatedSalary']])
```

```python
#Setting/Custom Index
df.set_index('User ID',inplace=True)
print(df)
```

```python
#Specific Rows
print(df.at[15810944,'Age'])
df.at[15668575,'Age']=50
print(df.loc[[15810944,15594041,15706071]])
print(df.loc[[15810944,15594041,15706071],["Gender",'Age','Purchased']])
print(df.loc[15810944:15728773])
```

```python
#conditional selection
print(df[df['Age']>45])
```

```python
print(df[(df['EstimatedSalary']>20000) & (df['EstimatedSalary']<=40000)])
```

```python
#Converting data type of columns
df['EstimatedSalary']=df['EstimatedSalary'].astype('float')
print(df.info())
```

```python
#Add new column
df['City']='Mumbai'
print(df.info())
print(df)
```

```python
#Drop columns or rows
df.drop('City',axis=1,inplace=True) #axis=1 for columns,axis=0 for rows
print(df)
```

```python
#Drop rows
df.drop(15810944,axis=0,inplace=True)
print(df)
```

```python
#Rename
df.rename(columns={'Gender':'Sex'},inplace=True)
print(df)
```

```python
#map()
df['Purchased']=df['Purchased'].map({0:'No',1:'Yes'})
print(df)
```

```python
#groupby()
print(df.groupby('Sex')['Age'].mean())
```

```python
#Value_counts()
print(df['Purchased'].value_counts())
df['Purchased'].value_counts().plot(kind='bar')
```

```python
#export Data
df_new=df[df['Age']>45]
print(df_new)
df_new.to_csv('new_data.csv')
df_new.to_excel('new_data.xlsx')
df_new.to_json('new_data.json')
```
