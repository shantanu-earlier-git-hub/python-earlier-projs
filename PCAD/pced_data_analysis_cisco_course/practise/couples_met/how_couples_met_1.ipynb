```python
import pandas as pd
import matplotlib.pyplot as plt

met_couples = pd.read_csv('how-couples-met.csv')
```

```python
corr_df = met_couples.corr()
corr_df
```

```python
plt.figure(figsize=(10,3))

plt.plot(corr_df.index, corr_df['online'])
plt.plot(corr_df.index, corr_df['decade'])
plt.plot(corr_df.index, corr_df['neighbors'])
plt.plot(corr_df.index, corr_df['through family'])

plt.legend(corr_df.columns, loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation='vertical')
plt.show()
```
