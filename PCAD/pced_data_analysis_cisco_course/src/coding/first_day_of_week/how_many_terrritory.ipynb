- How many territories show Friday, Saturday, Sunday, and Monday as the `first_day` of the week?


```python
import pandas as pd

df = pd.read_csv('first-day-of-week.csv')
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
criteria = df['first_day'].isin(['fri', 'sat', 'sun', 'mon'])
```

```python
df[criteria]
```

```python
df.drop(columns=['count'], inplace=True, axis=1)
```

```python
df[df['territory'].str.startswith(('C'))]
```

```python
new_criteria = df['first_day'].isin(['fri','sat','sun','mon'])
```

```python
territory_counts = df[new_criteria].count().reset_index()
```

```python
territory_counts.rename(columns={'index':'name', 0:'count'}, inplace=True)
```

```python
territory_counts.iloc[0].reset_index().transpose()
```
