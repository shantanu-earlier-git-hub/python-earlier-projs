```python
import pandas as pd

df = pd.read_csv('largest-islands.csv')
df.head()
```

- What islands are composed of multiple countries?
	- Hint: try `countries.str.contains(",")`

```python
multi_counties = df.query('countries.str.contains(",")')[['island', 'countries']].sort_values(by='island')
```
