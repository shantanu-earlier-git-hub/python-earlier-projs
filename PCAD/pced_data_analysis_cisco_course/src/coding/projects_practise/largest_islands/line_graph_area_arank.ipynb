```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df.head()
```

- Create a line graph with `area` on the y-axis and `rank` on the x-axis. The data should be ordered by `rank`, from largest to smallest.


```python
area_info_by_rank = df[['rank','area']].sort_values(by='rank', ascending=False)
```

```python
x = area_info_by_rank['rank']
y = area_info_by_rank['area']

plt.plot(x, y)
plt.xlabel('Rank')
plt.ylabel('Area')

plt.grid(True)
```
