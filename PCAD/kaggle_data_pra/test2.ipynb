```python
import pickle
from cmath import sqrt

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from fontTools.ttLib.woff2 import bboxFormat
```

```python
file = 'data_batch_1'

def get_data():
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes',errors='ignore')
    return dict
```

```python
all_data = get_data()
for keys in all_data.keys():
    print(str(keys))
```
