note that isnull() and isna() is moved to pd(pandas) directly to dataframe.

```python
def check_if_nan(value):
    if pd.isnull(value) or pd.isna(value):
        return False
    return True
```

```python
condition1 = df.apply(lambda df: check_if_nan(df['pre_movie_score']), axis=1)
```

```python
condition2 = df.apply(lambda df: check_if_nan(df['first_name']), axis=1)
```

```python
df[condition1]
```

drop all the rows as axis=0  where pre_movie_score col has NaN value

```python
df.dropna(axis=0, subset=['pre_movie_score'])
```

drop if any value is missing in a column

```python
df.dropna(axis='columns')
```

set subset to column names where to look for dropping the 'NaN' valued rows

```python
df.dropna(subset=['first_name','last_name','pre_movie_score','post_movie_score','age','sex'])
```

drop a row where all elements in a row are NaN, default axis = 0

```python
df.dropna(how='all')
```

for our dataset is axis = 1 nothing has been dropped

```python
df.dropna(how='all', axis=1)
```

thresh parameter checks if a row has how many non-NaN values if it is same or more than thresh value then it will be dropped

```python
df.dropna(thresh=5)
```

drop duplicates also get a subset plus has a keep option for keeping first or last record if found.

```python
df.drop_duplicates(keep='last', subset=['first_name','last_name'])
```

```python
df
```

### fillna

```python
df['pre_movie_score'] = df['pre_movie_score'].ffill()
df
```

```python
df['sex'] = df['sex'].fillna('M')
```

```python
df['post_movie_score'] = df['post_movie_score'].fillna(df['post_movie_score'].mean())
```

```python
df
```

Fill NA/NaN values by propagating the last valid observation to next valid.

```python
df['first_name'] = df['first_name'].ffill()
```

```python
df['age'] = df['age'].ffill()
```

```python
df
```

Fill NA/NaN values by using the next valid observation to fill the gap.

```python
df['last_name'] = df['last_name'].bfill()
```

```python
df
```

```python
df['pre_movie_score'] = df['pre_movie_score'].replace(to_replace=0, value=df['pre_movie_score'].mean())
```

```python
df
```

```python
df.to_csv("../../../data/movie_scores_cleaned.csv", index=False, header=True, sep=",")
```
