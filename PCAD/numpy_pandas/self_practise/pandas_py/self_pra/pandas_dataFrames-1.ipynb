```python
import numpy as np
import pandas as pd
```

#### dataframes consists of columns and rows, and it internally a bunch/group of Series Objects with same shared index(s)

```python
np.random.seed(101)
my_data = np.random.randint(0, 100, size=(4, 3))
```

```python
my_data
```

```python
my_index = ['ind', 'usa', 'mex', 'bra']
```

```python
my_cols_labels = ['q1', 'q2', 'q3']
```

#### data is enough

```python
pd.DataFrame(data=my_data)
```

#### add column names, pandas assign default index(s) as numerical serial values

```python
pd.DataFrame(data=my_data, columns=my_cols_labels)
```

#### just add labelled/named index remove cols

```python
df = pd.DataFrame(data=my_data, index=my_index)
df
```

#### add columns (they are just the labels here for columns) with data and index

```python
df = pd.DataFrame(data=my_data, index=my_index, columns=my_cols_labels)
df
```

```python
df.info()
```
