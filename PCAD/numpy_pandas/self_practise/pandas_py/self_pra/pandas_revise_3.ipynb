```python
import pandas as pd
import numpy as np
```

```python
heights = [5.1,6.2,7.1,5.0,5.2,5.4,20,-5,18,-3]
weights = [100,145,160,165,143,160,250,-200,200,180]
suger = np.linspace(100,110,10)
bmi = [h/w for h,w in zip(weights,heights)]
```

```python
h1 = pd.Series(data=heights, name="height")
w1 = pd.Series(data=weights, name="weight")
s1= pd.Series(data=suger, name="sugar")
bm1=pd.Series(data=bmi, name="bmi")
```

```python
df=pd.DataFrame({"height":h1, "weight":w1,"sugar":s1,"bmi":bm1})
```
