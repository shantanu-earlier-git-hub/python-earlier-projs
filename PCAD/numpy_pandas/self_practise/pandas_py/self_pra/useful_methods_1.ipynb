```python
def yelp(price):
    if price < 10:
        return '$'
    elif 10 <= price <= 20:
        return '$$'
    else:
        return '$$$'
```

please note here that apply function call a function ref will return a single value which will be applied to a column value in a row.

It doesn't return any series.

```python
def quality_tip(total_bill, tip):
    if tip/total_bill > 0.18:
        return "OK"
    else:
        return "Other"
```

```python
quality_tip(10.34,1.66)
```

```python
df[['total_bill','tip']].apply(lambda df:quality_tip(df['total_bill'],df['tip']), axis=1)
```

```python
np.vectorize(quality_tip)(df['total_bill'], df['tip'])
```
