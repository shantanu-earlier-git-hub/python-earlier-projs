```python
import numpy as np
import pandas as pd
```

```python
items = ['pen','pencil','book','notebook','pad']
prices = [10.00,5.34,2.30,1.23,.23]
```

```python
## is essentially a dictionary; but it can be formed with numpy arrays with named-index
## no need to give keyword arguments as it works fine with positionals, first as data and second as index
#shop1 = pd.Series(data=prices,index=items)
shop1 = pd.Series(prices,items)
```

```python
shop1
```

```python
## can be accessed with named index
shop1['pen']
```

```python
# shop1[0] -> deprecated
## use iloc[] for accessing items using numeric inbuilt index
shop1.iloc[0]
```

```python
ages = {'sam': 10, 'sakey':20, 'mike':30,'doc':12}
```

```python
## series object can be created using dict, keys as named index and values as values...
## So dataframes are actually the array of series/ dictionaries.
age_series = pd.Series(ages, name='age')
```

```python
age_series
```

```python
shop1_items = ['pen','pencil','book','binder','pad']
shop1_prices = [10.00,5.34,2.30,1.23,.23]
shop1 = pd.Series(data=shop1_prices,index=shop1_items, name='price')

shop2_items = ['pen','pencil','textbook','notebook','pad']
shop2_prices = [10.00,5.34,2.30,1.23,.23]
shop2 = pd.Series(data=shop2_prices,index=shop2_items, name='price')
```

```python
shop1
```

```python
shop2
```

```python
# if an item not present pandas doesn't know what to do with such items and hence that price will be NAN
# for those.

shop1 + shop2
```

```python
## fill the value for not available items with 0.00 and then add
shop1.add(shop2, fill_value=0.00)
```
