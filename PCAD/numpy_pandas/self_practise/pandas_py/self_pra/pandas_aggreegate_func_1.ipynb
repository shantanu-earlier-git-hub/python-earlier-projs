```python
import numpy as np
import pandas as pd

df = pd.read_csv("../../../../data/mpg.csv")
```

```python
df.head()
```

```python
df = df.drop(['name','horsepower'], axis=1)
```

```python
df.columns
```

```python
df.info()
```

```python
df.groupby(['model_year','cylinders']).mean().head()
```

aggregate functions can work on single and multiple columns, and multiple functions for column selections use dict.

```python
agg_df = df.aggregate({'weight':['mean','max'], 'mpg':['median','min']})
```

```python
agg_df
```

agg on groupby with one col selected

```python
df.groupby(['model_year','cylinders'])['mpg'].agg(['mean','max','min','std','var'])
```

abbreviated agg() of aggregation() use [] for function applied for entire df

```python
df.agg(['std','mean','max','min','var']).transpose()
```

```python
df.agg(['std','mean','max','min','var'])[['mpg','displacement','weight']].transpose()
```
