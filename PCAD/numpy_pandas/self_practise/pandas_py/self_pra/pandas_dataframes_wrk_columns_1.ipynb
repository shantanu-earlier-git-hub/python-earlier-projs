### Pandas Columns

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../../../../data/tips.csv')
```

```python
df.columns
```

```python
df['total_bill'].head()
```

if column present it recalculate the column values and overwrite those

```python
df['price_per_person'] = df['total_bill'] / df['size']
```

```python
df.head()
```

```python
df['price_per_person'] = np.round(a = df['total_bill'] / df['size'], decimals=2)
```

```python
df.head()
```

adding a new column by calculation 'tip_percentage'

```python
df['tip_percentage'] = np.round((df['tip'] / df['total_bill']) * 100, decimals=2)
```

```python
df.head()
```

using evaluate() function

```python
df['tip_percentage'] = df.eval('(tip / total_bill) * 100')
```

```python
df.head()
```

```python
df['tip_percentage'] = np.round(df.eval('(tip / total_bill) * 100'), 2)
```

```python
df.head()
```

#### df.drop() for dropping a row or a column

```python
df.drop('tip_percentage', axis=1)
```
