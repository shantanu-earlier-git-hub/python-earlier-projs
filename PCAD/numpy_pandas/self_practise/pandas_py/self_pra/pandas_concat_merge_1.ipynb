```python
import uuid
def get_uuid():
    yield str(uuid.uuid4())

def gen_uuids():
    uuids = []
    for i in range(5):
        uuid_str = get_uuid().__next__()
        uuids.append(uuid_str)

uuids
```

```python
books = {
    '_id' : uuids,
    'names' :['python','java','ruby','rails','numpy'],
    'prices' : np.linspace(20,30,5)
}
```

```python
books_df = pd.DataFrame(books)
books_df
```

```python
authors = {
    '_id' : uuids,
    'names' :['','java','ruby','rails','numpy'],
    'authors' : ['tom','mike','bob','nikki','shawn'],
}
```

```python
authors_df = pd.DataFrame(authors)
authors_df
```

merging will work only both data frames has matched values. if one of them doesn't have it then apply left or right join to catch having one ...

apply outer join if one value present and other doesn't ... but pandas will replace i other relative values as NaN in merged dataframe


inner...
when _id and names ==  _id and names

```python
pd.merge(books_df, authors_df, on=['_id','names'], how='inner')
```

left , include left

_id and names +==  _id and names

```python
pd.merge(books_df, authors_df, on=['_id','names'], how='left')
```

right, include right

_id and names ==+  _id and names

```python
pd.merge(books_df, authors_df, on=['_id','names'], how='right')
```

both, include both no matter present, cartesian product

_id and names == _id and names

```python
merged_outer = pd.merge(books_df, authors_df, on=['_id','names'], how='outer')
```
