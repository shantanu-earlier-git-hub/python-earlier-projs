```python
import pandas as pd

df_territory = pd.read_csv("first-day-of-week.csv")
```

```python
df_territory.head()
```

```python
df_population = pd.read_csv("population.csv")
```

```python
df_population.head()
```

```python
df_tp_merged = df_territory.merge(df_population, on="alpha3", how='left')
```

```python
df_tp_merged
```

```python
df_regions = pd.read_csv('four-regions.csv')
```

```python
df_regions.head()
```

```python
final_df = df_tp_merged.merge(df_regions, on="alpha3", how='left')
```

```python
final_df.columns
```

```python
col_map = {
    'territory': 'country',
    'four_regions': 'region'
}
final_df.rename(columns=col_map, inplace=True)
```

```python
final_df.head()
```

```python
final_df[final_df['population'].isna()]
```

```python
final_df.loc[final_df['population'].isna(), 'population'] = 0
```

```python
region_mode = final_df['region'].mode()
```

```python
final_df['region'].astype(str)
```

```python
final_df.loc[final_df['region'].isna(), 'region'] = region_mode
```

```python
final_df.loc[final_df['region'].isna(), "region"] = 'america'
```

```python
final_df
```

```python
final_df.loc[final_df['alpha3'].isna(), "alpha3"] = 'AME'
```
